```python
import pandas as pd
import numpy as np
import os
from bs4 import BeautifulSoup
import tensorflow
import string
from copy import copy
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from collections import Counter
```

```python
def preprocess_nlp(text,remove_html=True,lower=True,remove_punct=True):
    """remove html tags from text"""
    if remove_html==True:
        soup = BeautifulSoup(text, "html.parser")
        text = soup.get_text(separator=" ")
    if lower==True:
        text=text.lower()
    if remove_punct==True:
        table = str.maketrans(string.punctuation,' ' * len(string.punctuation))
        text=text.translate(table).strip()
    return text
```

```python
directory=os.getcwd().replace('notebooks','data\\raw\\')
data={'biology':pd.read_csv(f'{directory}biology.csv'),
     'cooking':pd.read_csv(f'{directory}cooking.csv'),
     'crypto':pd.read_csv(f'{directory}crypto.csv'),
     'diy':pd.read_csv(f'{directory}diy.csv'),
     'robotics':pd.read_csv(f'{directory}robotics.csv'),
     'travel':pd.read_csv(f'{directory}travel.csv')}
```

```python
travel_df=copy(data['travel'])#.set_index('id')
travel_df['content']=data['travel']['content'].map(preprocess_nlp)
travel_df['title']=data['travel']['title'].map(preprocess_nlp)
travel_df
```

```python
travel_df.set_index('id',inplace=True)
```

```python
def count_unique_words(df_series):
    results = Counter()
    df_series.str.split().apply(results.update)
    return(len(results.keys()))
unique_words_title=count_unique_words(travel_df['title'])
unique_words_content=count_unique_words(travel_df['content'])
```

```python
X=travel_df[['title','content']]
y=travel_df['tags']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
tk = Tokenizer(num_words = unique_words_title)
tk.fit_on_texts(X_train)
X_train = tk.texts_to_sequences(X_train)
X_test = tk.texts_to_sequences(X_test)
X_train
```
